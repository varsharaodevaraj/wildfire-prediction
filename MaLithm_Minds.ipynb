```python
noaa_df = pd.read_csv('uttarakhand_weather_2010_2023.csv')
```

```python
noaa_df.head()
```

pivot this dataset

```python
noaa_copy = noaa_df[["date","datatype","value"]].copy()
```

```python
noaa_copy.head()
```

```python
noaa_copy["date"].dtype
```

```python
noaa_copy.info()
```

```python
noaa_copy["date"] = pd.to_datetime(noaa_copy["date"])
```

```python
noaa_copy["date"].dtype
```

```python
noaa_copy.info()
```

```python
noaa_copy.isnull().sum()
```

```python
noaa_copy = noaa_copy.pivot_table(
    index="date",
    columns="datatype",
    values="value",
    aggfunc="mean",

).reset_index()
```

```python
noaa_copy.head()
```

```python
noaa_original = noaa_copy[["date","PRCP","SNWD","TMAX","TMIN"]].copy()
```

```python
noaa_original.head()
```

```python
noaa_original = noaa_original.ffill().sort_values(by="date")
```

```python
noaa_original
```

```python
noaa_original = noaa_original.bfill().sort_values(by="date")
```

```python
noaa_original
```
